```python
##QUESTION1

text = input("Enter string: ")
count = 0

for ch in text:
    if ch.isupper():
        count += 1

print("Capital letters:", count)
```

```python
##QUESTION2


text = input("Enter string: ")
count = 0

for ch in text.lower():
    if ch in "aeiou":
        count += 1

print("Total vowels:", count)
```

```python
##QUESTION3

sentence = input("Enter sentence: ")
words = sentence.split()

for w in words:
    print(w)
```

```python
##QUESTION5

text = input("Enter string: ").upper()

for ch in sorted(set(text)):
    if ch.isalpha():
        print(text.count(ch), ch, sep="")
```

```python
##QUESTION6

sentence = input("Enter sentence: ")
words = sentence.split()

unique_words = set(words)

print("Unique words count:", len(unique_words))
```

```python
##QUESTION7

n = int(input("Enter number of fruits: "))

s1 = set()
s2 = set()

print("Enter fruits for set 1:")
for i in range(n):
    s1.add(input())

print("Enter fruits for set 2:")
for i in range(n):
    s2.add(input())

print("Common fruits:", s1 & s2)
print("Only in s1:", s1 - s2)
print("Total fruits count:", len(s1 | s2))
```

```python
##QUESTION8


S1 = {"Red", "yellow", "orange", "blue"}
S2 = {"violet", "blue", "purple"}

print("Union:", S1 | S2)
print("Intersection:", S1 & S2)
print("Only in S1:", S1 - S2)
print("Only in S2:", S2 - S1)
print("Symmetric Difference:", S1 ^ S2)
```
