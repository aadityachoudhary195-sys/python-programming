```python
###question 1a



with open("name.txt", "w") as f:
    n = int(input("Enter number of names: "))
    for i in range(n):
        name = input()
        f.write(name + "\n")


```

```python
##question 1c



count = 0
with open("name.txt", "r") as f:
    for line in f:
        if line.strip()[0].lower() in "aeiou":
            count += 1
print("Names starting with vowel:", count)
```

```python
##question 1d


longest = ""
with open("name.txt", "r") as f:
    for line in f:
        name = line.strip()
        if len(name) > len(longest):
            longest = name

print("Longest name:", longest)
```

```python
##question 2a



with open("numbers.txt", "r") as f:
    nums = [int(x.strip()) for x in f]

print("Maximum:", max(nums))
```

```python
##question 2b


count = 0

with open("numbers.txt", "r") as f:
    numbers = [int(x.strip()) for x in f]

for num in numbers:
    if num > 100:
        count += 1

print("Numbers greater than 100:", count)
```

```python
##question4


total_area = 0

with open("city.txt", "r") as f:
    for line in f:
        data = line.split()
        total_area += float(data[2])
        
print("Total area:", total_area)
```
