```python
##QUESTION1



def find_max_min(numbers):
    maximum = numbers[0]
    minimum = numbers[0]
    
    for num in numbers:
        if num > maximum:
            maximum = num
        if num < minimum:
            minimum = num
    
    return maximum, minimum

nums = list(map(int, input("Enter numbers: ").split()))
result = find_max_min(nums)
print("Maximum:", result[0])
print("Minimum:", result[1])
```

```python
##QUESTION2



def sum_of_cubes(n):
    total = 0
    for i in range(1, n):
        total += i ** 3
    return total

n = int(input("Enter number: "))
print("Sum of cubes:", sum_of_cubes(n))
```

```python
##QUESTION3




def print_numbers(n):
    if n == 0:
        return
    print_numbers(n - 1)
    print(n)

n = int(input("Enter number: "))
print_numbers(n)
```

```python
##QUESTION4


def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

n = int(input("Enter number of terms: "))

for i in range(n):
    print(fibonacci(i), end=" ")
```

```python
##QUESTION5



volume = lambda r, h: (1/3) * 3.14 * r * r * h

r = float(input("Enter radius: "))
h = float(input("Enter height: "))

print("Volume of cone:", volume(r, h))
```
